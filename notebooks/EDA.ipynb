UNDERSTANDING THE IMPORTANCE OF A FEATURE AND HANDLING MISSING VALUES (2 Approaches)

1) Data Cleaning Basic

```python
from sklearn.impute import KNNImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer



# Copy original dataset for both approaches
df_basic = df_uci.copy()


# ==================================
# 🟢 APPROACH 1: BASIC CLEANING
# ==================================

# Impute missing values using mean/median/mode
basic_impute_dict = {
    'trestbps': df_basic['trestbps'].median(),
    'chol': df_basic['chol'].median(),
    'thalch': df_basic['thalch'].median(),
    'oldpeak': df_basic['oldpeak'].median(),
    'fbs': df_basic['fbs'].mode()[0],
    'restecg': df_basic['restecg'].mode()[0],
    'exang': df_basic['exang'].mode()[0]
}

# Apply imputation
df_basic.fillna(basic_impute_dict, inplace=True)

# Drop highly missing columns (optional)
df_basic.drop(columns=['ca', 'thal', 'slope'], inplace=True)  

# Save cleaned dataset
df_basic.to_csv("heart_disease_cleaned_basic.csv", index=False)
```

Columns ca, thal, slope are dropped (>100 missing records)

Remaining missing data are handled using MMM imputation


2) Advanced Data Cleaning based on Importance

```python
from sklearn.preprocessing import LabelEncoder

df_advanced = df_uci.copy()
df_advanced.astype(str)
knn_cols = ["ca", "thal", "slope"]

# Convert categorical values to numeric before KNN
label_encoders = {}
for col in knn_cols:
    df_advanced[col] = df_advanced[col].astype(str)  # Ensure string type
    le = LabelEncoder()
    df_advanced[col] = le.fit_transform(df_advanced[col])
    label_encoders[col] = le  # Store encoder for inverse transformation if needed

```

```python
# Columns for different imputation techniques
knn_cols = ['ca', 'thal', 'slope']  # Highly missing but important
mice_cols = ['trestbps', 'chol', 'thalch', 'oldpeak']  # Continuous with patterns
mode_cols = ['fbs', 'restecg', 'exang']  # Categorical

# ==================================
# 🔵 APPROACH 2: ADVANCED CLEANING
# ==================================

# KNN Imputation for highly missing but important features
knn_imputer = KNNImputer(n_neighbors=5)
df_advanced[knn_cols] = knn_imputer.fit_transform(df_advanced[knn_cols])

# Round categorical values back to integers
df_advanced[knn_cols] = df_advanced[knn_cols].round(0).astype(int)

# MICE (Iterative Imputer) for continuous data
mice_imputer = IterativeImputer(max_iter=10, random_state=42)
df_advanced[mice_cols] = mice_imputer.fit_transform(df_advanced[mice_cols])

# Mode Imputation for categorical variables
for col in mode_cols:
    df_advanced[col].fillna(df_advanced[col].mode()[0], inplace=True)

# Save cleaned dataset
df_advanced.to_csv("heart_disease_cleaned_advanced.csv", index=False)
```

MICE imputation for continuous features (trestbps, chol, thalch, oldpeak).

Mode imputation for categorical features (fbs, restecg, exang).

KNN imputation for thal, ca, and slope

3) Optimized Advanced using MICE for ca, thal, slope

```python
from sklearn.impute import KNNImputer, SimpleImputer
df_advanced_1 = df_uci.copy()
df_advanced_1.astype(str)
mice_cols = ["ca", "thal", "slope"]

# Convert categorical values to numeric before KNN
label_encoders = {}
for col in mice_cols:
    df_advanced_1[col] = df_advanced_1[col].astype(str)  # Ensure string type
    le = LabelEncoder()
    df_advanced_1[col] = le.fit_transform(df_advanced_1[col])
    label_encoders[col] = le  # Store encoder for inverse transformation if needed

# 1️⃣ **MICE Imputation for `ca`, `thal`, `slope` (Highly missing but important)**
mice_imputer = IterativeImputer(max_iter=10, random_state=42)
mice_cols = ["ca", "thal", "slope"]
df_advanced_1[mice_cols] = mice_imputer.fit_transform(df_advanced_1[mice_cols])

# 2️⃣ **KNN Imputation for `trestbps`, `chol`, `thalch`, `oldpeak` (Continuous features with patterns)**
knn_imputer = KNNImputer(n_neighbors=5)
knn_cols = ["trestbps", "chol", "thalch", "oldpeak"]
df_advanced_1[knn_cols] = knn_imputer.fit_transform(df_advanced_1[knn_cols])

# 3️⃣ **Mode Imputation for `fbs`, `restecg`, `exang` (Categorical features)**
mode_imputer = SimpleImputer(strategy="most_frequent")
mode_cols = ["fbs", "restecg", "exang"]
df_advanced_1[mode_cols] = mode_imputer.fit_transform(df_advanced_1[mode_cols])

# Save the cleaned dataset
df_advanced_1.to_csv("_optimized_advanced_heart_disease_cleaned.csv", index=False)


```

✅ MICE (Multiple Imputation by Chained Equations) for ca, thal, slope

These categorical features have high missing values but are important for heart disease prediction.
MICE is good for categorical features because it models missing values based on other feature relationships iteratively.

✅ KNN Imputation for trestbps, chol, thalch, oldpeak

These are continuous features, and KNN is effective for finding similar patterns in existing data to fill in missing values.
KNN uses similar neighbors, making it better for structured numerical data.

✅ Mode Imputation for fbs, restecg, exang

These are categorical variables with low missing values, and the mode (most frequent value) is a simple but effective way to handle them.
